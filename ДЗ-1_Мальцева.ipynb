## Задание 1: (условия if-elif-else)

В ячейке ниже - код "определителя усталости". Давайте его улучшим:

- добавьте условие для случаев,если пользователь вводит число больше 10

- добавьте еще одно или два условия, чтобы разделение на категории усталости было еще точнее

Подсказка: может помочь нарисовать себе схему условий на бумаге


````python
# определитель усталости

a = float(input('по шкале от 0 до 10, напишите, насколько Вы устали: '))
if a < 0.7:
    print('Кажется, Вы совсем не устали')
elif 0.7 <= a <= 4.8:
    print('Вы немного устали')
# Ваш код дальше:
elif 4.9 <= a <= 7.4:
    print('У Вас средний уровень усталости')
elif 7.5 <= a <= 8.9:
    print('Повышенный уровень усталости - Вам следует отдохнуть!')
elif 9 <= a <= 10:
    print('Вы очень устали. Срочно спать!')
else:
    print('Похоже, Вы чертовски сильно устали. Рекомендуется отпуск!')
````

### Задача 2 (списки и циклы)
Робот, выбирающий пользователю коктейль:

Представьте, что Вы разрабатываете (фрагмент) электронного помощника для ресторана. 

У Вас есть список коктейлей, где каждый коктейль представлен в виде списка ингридиентов (у нас получается список списков). Программа рандомным образом должна выбрать один из коктейлей (т.е. один из списков в общем списке) и вывести на экран сообщение: "Сегодня в вашем коктейле будет: " и сам список ингридиентов

Но есть одно условие: в некоторые из списков по ошибке попал "лосось". Строку "лосось" нужно убрать *(здесь вам поможет один метод работы со списками)* ,но так как лосось есть не везде, нужно предварительно проверить его наличие в списке.

````python
import random

cocktails = [
    ["мартини","грейпфрутовый сок","жасмин","тоник","лосось"],
    ["клубника","какао","мята","марсала"],
    ["водка","томатный сок","лимонный сок","вустерширский соус","черный перец","сельдерей","лосось"],
    ["джин","вермут","ликер мараскино","апельсины","коктейльная вишня","лосось"],
    ["ром","авокадо","сахарный сироп","сливки","лимонный сок","лед"],
    ["красный вермут","тоник","апельсины","лосось"],
    ["только чай"]
        ]

#Ваш код
for i in cocktails:
    if "лосось" in i:
        i.remove('лосось')
print("Сегодня в вашем коктейле будет:", random.choice(cocktails))
````

````python
import random

# cocktails = [
#     ["мартини","грейпфрутовый сок","лосось","жасмин","тоник"],
#     ["клубника","какао","мята","марсала"],
#     ["водка","томатный сок","лимонный сок","вустерширский соус","черный перец","сельдерей","лосось"],
#     ["джин","вермут","ликер мараскино","апельсины","коктейльная вишня","лосось"],
#     ["ром","авокадо","сахарный сироп","сливки","лимонный сок","лед"],
#     ["красный вермут","тоник","апельсины","лосось"],
#     ["только чай"]
#         ]

#Ваш код
juices = ['грейпрутовый сок','лимонный сок',"томатный сок","апельсиновый сок","яблочный сок"]
fruits = ["клубника","апельсин","вишня","авокадо","лайм","яблоко","виноград"]
alcool = ["мартини","текила","марсала","водка","ром","коньяк","хреновуха","виски","джин","вермут","ликер"]
soda = ["кока-кола","тоник","спрайт","газировка","пепси"]
spices = ["сельдерей","жасмин","вустершский соус","черный перец","мята","сахарный сироп","корица","сливки","какао","шоколад"]
tea = ["красный чай","зеленый чай","черный чай","травяной чай","фруктовый чай"]
coffee = ["капучино","раф","эспрессо","латте","маккиато","моккачино","американо","какао","горячий шоколад"]

name = input("Добрый вечер! Как я могу к Вам обращаться? ")
# name.upper()
drink = input('Что бы Вы хотели выпить? ')
if drink in alcool:
    cocktail = [drink, random.choice(juices), random.choice(fruits), random.choice(soda), random.choice(spices)]
    print("Предлагаю Вам коктейль", name.upper(), "ВСЁ!:",cocktail)
elif (drink in juices)|(drink in soda):
    agreement = input("Сегодня Вы против алкоголя? ")
    if agreement == "нет":
        cocktail = [drink, random.choice(alcool), random.choice(fruits), random.choice(soda), random.choice(spices)]
        print('Предлагаю Вам коктейль ', name.upper(), 'ВСЁ!:',cocktail)    
    else:
        print("Что ж, желаю приятного вечера,", name, "!")
elif (drink in tea)|(drink in coffee):
    extra = input("Хотите добавить специй и фруктов? ")
    if extra == "да":
        fruits.remove("авокадо")
        if (drink == "какао")|(drink == "горячий шоколад"):
            spices.remove("шоколад")
            spices.remove("какао")
            
        print("Сегодня для Вас,",name, ":", "\n", drink, "\n", random.choice(fruits), "\n", random.choice(spices))
    else:
        print("Благодарю Вас и желаю приятного вечера,", name, "!")
````
